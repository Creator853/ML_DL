```python
#Import the necessary libraries 
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn.decomposition import PCA
from sklearn.metrics import mean_squared_error
```

```python
#Load in the data and check the shapes
train = pd.read_csv("train.csv")
test = pd.read_csv("test.csv")
print("Shapes --> \n Train: ", train.shape ,"\n Test: ", test.shape)
```

```python
train.head()
```

```python
#Separation of independent and dependent variables
train_x = train.drop(columns=["Item_Outlet_Sales"])
train_y = train["Item_Outlet_Sales"]
test_x = test.drop(columns=["Item_Outlet_Sales"])
test_y = test["Item_Outlet_Sales"]
```

```python
#Let's re-confirm the number of dimensions on training data.
print('\nTraining model with {} dimensions.'.format(train_x.shape[1]))
```

```python
#Initilisation of Model Object
model = LinearRegression()
model.fit(train_x, train_y) 
```

```python
#Prediction on train and test data
print("Train: \n ", model.predict(train_x), " \n Test: \n ", model.predict(test_x))
```

```python
#Accuracy on train and test data
print("Accuracy on train data: {} ".format(mean_squared_error(train_y, model.predict(train_x))**0.5))
print("Accuracy on test data: {}".format(mean_squared_error(test_y, model.predict(test_x))**0.5))
```

Till now we were considering the simple linear regression model now let's try using PCA which stands for 
PRINCIPAL COMPONENT ANALYSIS    
As the name suggests we'll be performing a reduction in dimensionality of the dataset to find the dimensions due to which o/p is affected the most.

```python
#Initialisation of PCA object
model_pca = PCA(n_components=12)
train1_x, test1_x = model_pca.fit_transform(train_x), model_pca.fit_transform(test_x)

#Let's find out the dimensions of new training and testing data
print("\nTraining: {}".format(train1_x.shape[1]),"\n Testing: {}".format(test1_x.shape[1]))
```

```python
#Now let's fit the model with new training data.
model1 = LinearRegression()
model1.fit(train1_x, train_y)
```

```python
#Now lets make predictions on new training and testing data.
print("Predictions on Train: \n {}".format(model1.predict(train1_x)))
print("\nPredictions on Test: \n {}".format(model1.predict(test1_x)))
```

```python
#Analysis of performance metrics on both train and test data
print("RMSE for Train: {}".format(mean_squared_error(train_y, model1.predict(train1_x))**0.5))
print("RMSE for Test: {}".format(mean_squared_error(test_y, model1.predict(test1_x))**0.5))
```

Haalaanki (Although) there hasn't been much improvement in the RMSE for this dataset but this is a very good method to find and plot the most important independent variables which should be used while predicting the outcome.
